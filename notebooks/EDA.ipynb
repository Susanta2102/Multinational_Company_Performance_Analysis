# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load cleaned dataset
data_path = "../data/processed/Cleaned_Multinational_Company_Performance.csv"
df = pd.read_csv(data_path)

# Display dataset information
print(df.info())
print(df.head())
```

```python
# Display summary statistics
print(df.describe())
```

```python
# Check for missing values
print("Missing Values per Column:")
print(df.isnull().sum())

# Check for duplicates
print("Duplicate Rows Count:", df.duplicated().sum())
```

```python
# Aggregate revenue by country
revenue_by_country = df.groupby('Country')['Revenue'].sum().reset_index()

# Visualize revenue distribution
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Revenue', data=revenue_by_country, palette='viridis')
plt.title('Revenue by Country')
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Aggregate units sold by product category
units_by_category = df.groupby('Product Category')['Units Sold'].sum().reset_index()

# Visualize units sold
plt.figure(figsize=(8, 5))
sns.barplot(x='Product Category', y='Units Sold', data=units_by_category, palette='coolwarm')
plt.title('Units Sold by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.show()
```
